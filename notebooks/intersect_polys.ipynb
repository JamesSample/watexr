```python
%matplotlib inline
import nivapy3 as nivapy
import pandas as pd
import geopandas as gpd
```

```python
# Read data
kom_path = '../gis/vector/NO_Kommuner_pol.shp'
cat_path = '../gis/vector/morsa_catchmentBoundary.shp'
kom_gdf = gpd.read_file(kom_path)
cat_gdf = gpd.read_file(cat_path).to_crs(kom_gdf.crs)
```

```python
print (len(kom_gdf))
kom_gdf.plot()
```

```python
print (len(cat_gdf))
cat_gdf.plot()
```

```python
# Intersect
gdf = nivapy.spatial.spatial_overlays(kom_gdf, cat_gdf)
gdf
```

```python
# Plot
gdf.plot()
```

```python
# Save as shp
out_shp = r'../gis/vector/morsa_kommuner.shp'
gdf.to_file(out_shp)
```

```python
nivapy.spatial.plot_shapefile(out_shp)
```
